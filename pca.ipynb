```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
x= [2.4, 0.6, 2.1, 2, 3, 2.5, 1.9, 1.1, 1.5, 1.2]
y= [2.5, 0.7, 2.9, 2.2, 3.0, 2.3, 2.0, 1.1, 1.6, 0.8]
```

```python
x= np.array(x)
y= np.array(y)
plt.scatter(x,y)
```

```python
x_m= np.mean(x)
y_m= np.mean(y)

x= x-x_m
y= y-y_m

plt.scatter(x,y)
```

```python
c= np.cov(x,y)
c
```

```python
from numpy import linalg as LA
w, v = LA.eig(c)
print("w:",w)
print()
print("v:",v)
```

```python
p1= v[:,1]
p2= v[:,0]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.plot([-2*p1[0],2*p1[0]],[-2*p1[1],2*p1[1]])
```

```python
plt.scatter(x,y)
plt.plot([-2*p1[0],2*p1[0]],[-2*p1[1],2*p1[1]])
plt.plot([0,p2[0]],[0,p2[1]])
```
